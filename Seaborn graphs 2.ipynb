## Histplot

```python
penguins = sns.load_dataset('penguins').dropna()
penguins.shape
```

```python
sns.set_style('darkgrid')
penguins.head()
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins);
```

```python
sns.histplot(y = 'bill_length_mm', data = penguins);
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins,
            kde = True);
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins, bins= 20);
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins
            , binrange= (50,60));
```

#### stat argument

```python
sns.histplot(x = 'bill_length_mm', data = penguins,
             stat= 'count' );
```

```python
# density = AREA  of histogram sums to 1
sns.histplot(x = 'bill_length_mm', data = penguins,stat = 'density');
```

```python
# probability = Height of histogram bars to one
sns.histplot(x = 'bill_length_mm', data = penguins,stat = 'probability');
```

```python
# cumulative 
sns.histplot(x = 'bill_length_mm', data = penguins,stat = 'density'
            ,fill = False, element= 'step');
```

```python

sns.histplot(x = 'bill_length_mm', data = penguins,stat = 'density'
            ,fill = False, element= 'step', cumulative= True);
```

```python
# element
sns.histplot(x = 'bill_length_mm', data = penguins, hue = 'species');
```

```python
# element
sns.histplot(x = 'bill_length_mm', data = penguins,
             element= 'step', hue = 'species');
```

```python
# element
sns.histplot(x = 'bill_length_mm', data = penguins,
             element= 'poly', hue = 'species');
```

```python
# Element
sns.histplot(x = 'bill_length_mm', data = penguins,
             multiple= 'stack', hue = 'species');
```
