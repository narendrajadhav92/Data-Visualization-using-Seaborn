```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('mtcars.csv')
df.head(10)
```

```python
#carb=number of carborators - device for supplying spark ignition
#cyl -cylinders
#gear -gears
```

```python
df.info()  # gives information of dataframe
```

```python
df.describe()
```

```python
df.shape
```

```python
#barplot 
result=sns.barplot(data=df,
                  x="cyl",
                  y="mpg")
plt.show()   
```

```python
result=sns.barplot(data=df,
                  x="cyl",
                  y="mpg",
                  color='r')
plt.show()     # to show the bar plot graph
```

```python
sns.barplot(data=df,
           x="am",
           y="mpg",
           palette="tab10")
plt.show()
```

```python
sns.barplot(data=df,
           x="am",
           y="mpg",
           palette="Paired")
plt.show()
```

```python
sns.barplot(data=df,
           x="am",
           y="mpg",
           palette="Set2")
plt.show()
```

```python
sns.barplot(data=df,
           x="gear",   #no of gears
           y="hp",   #horse power
           palette="rocket")
plt.show()
```

```python
#creating countplot
#returns count for each category in the data
```

```python
sns.countplot(data=df,
           x="cyl",
              
              
              
              
           palette="tab10")


#there is no y axis here since y axis just shows the count 
```

```python
sns.countplot(data=df,
           y="gear",
           palette="Set2") # no x axis
```

```python
sns.countplot(data=df,
           x="gear",   #no of gears
           hue="cyl",   #horse power
           palette="rocket")
#for each gear hue tells the no of cylinders
```

##### distplot- creates distribution of continuous data: age, height, weight,stock price, movie ticket, blood pressure

```python
sns.distplot(df.mpg, #dataframe.columnname
           bins=10,
           color='g')
```

```python
df.corr()
```

```python
#create a heatmap _visualize matrix used for correlations

sns.heatmap(df.corr(),
           annot=True,
           cbar=True, # colorbar if cbar=false range will not be shown
           linewidths=4)
```

```python
sns.heatmap(df.corr(),
           annot=False,
           cbar=False, # colorbar if cbar=false range will not be shown
           linewidths=4)
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.columns
```

```python
iris.shape
```

```python
iris['species'].value_counts()
```

```python
#scatterplot - shows relationship
#scatter plot -> outlier detection
#box-whisker plot -> outlier detection
```

```python
sns.scatterplot(data=iris,
               x='sepal_length',
               y='petal_length', color="g")
```

```python
sns.scatterplot(data=iris,
               x='sepal_length',
               y='petal_length', hue="species")
```

```python
#pairplots= relationship of each variable
sns.pairplot(iris, palette="Set2")
```

```python
sns.pairplot(iris,
            hue="species",
            palette="Set1")
```

```python
sns.lmplot(x="petal_length",
          y="petal_width",
          data=iris)


#the line is regression line
#we can use hue and put markers
```

```python
sns.lmplot(x="petal_length",
          y="petal_width",
          hue="species",
          markers=["o","*","x"],
          data= iris)
```

```python
#boxplot - used for outlier analysis
```

```python
sns.boxplot(data=iris,
           x="species",
           y="sepal_width")
```

```python
tips=sns.load_dataset("tips")
```

```python
sns.catplot(data=tips,
            x="smoker", 
            y="tip", 
            order=["No", "Yes"])
```

```python
# the jitter parameter controls the magnitude of jitter or disables is altogether:
sns.catplot(data=tips,
           x="day",
           y="total_bill",
           jitter=False)
```
